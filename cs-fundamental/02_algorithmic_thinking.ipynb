# 🛠️ **Hands-on Lab: Time & Space Optimization Mission**

### **🚀 Station 1: The "Needle in a Haystack" (Meb Search)**
**โจทย์:** บริษัทมีนิยายอยู่ 10 ล้านเล่ม (เรียง ID ไว้แล้ว) การค้นหาแบบเส้นตรง (Linear) ทำให้ Server ทำงานหนักมาก จงเขียน Binary Search เพื่อลดเวลาให้เหลือแค่เสี้ยววินาที

```python
import time

# สร้างข้อมูลจำลอง (ID นิยาย 10 ล้านเล่ม)
print("Generating 10M books... (Please wait)")
books_db = list(range(1, 10000001))
target_id = 9999999 # ค้นหาตัวเกือบสุดท้าย (Worst Case)

# --- 1. Linear Search O(N) ---
def linear_search(data, target):
    steps = 0
    for item in data:
        steps += 1
        if item == target:
            return steps
    return steps

# --- 2. Binary Search O(log N) ---
def binary_search(data, target):
    steps = 0
    left = 0
    right = len(data) - 1

    while left <= right:
        steps += 1
        mid = (left + right) // 2

        # ✅ Mission 1: เติม Logic ของ Binary Search ให้สมบูรณ์
        if data[mid] == target:
            return steps
        elif data[mid] < target:
            # ต้องแก้ left หรือ right อย่างไร?
            left = mid + 1
            pass
        else:
            # ต้องแก้ left หรือ right อย่างไร?
            right = mid - 1
            pass

    return steps

# ทดสอบความเร็ว
print("\n--- Searching O(N) vs O(log N) ---")
start = time.time()
steps_n = linear_search(books_db, target_id)
print(f"Linear: Found in {steps_n} steps ({time.time() - start:.5f} sec)")

# เมื่อเติม Code เสร็จ ลองรันเพื่อดูจำนวน Steps ของ Binary นะครับ!
steps_log = binary_search(books_db, target_id)
print(f"Binary: Found in {steps_log} steps")
```

### **🕵️‍♂️ Station 2: The "Fraud Blocker"**
**โจทย์:** คุณมี Blacklist บัญชีมิจฉาชีพ 1 ล้านบัญชี ระบบต้องเช็คว่าบัญชีที่กำลังจะโอนเงิน (test_account) อยู่ในแบล็คลิสต์ไหม ปัจจุบัน Dev ใช้ List ทำให้ระบบโอนเงินค้าง!

```python
# สร้าง Blacklist 1 ล้านรายการ
blacklist_list = list(range(1, 1000001))

# ✅ Mission 2: แปลง blacklist_list ให้เป็น Data Structure ที่ค้นหาได้ใน O(1)
blacklist_fast = set(blacklist_list)

test_account = 999999

print("\n--- Lookup O(N) vs O(1) ---")

# แบบเก่า (List)
start = time.time()
is_fraud = test_account in blacklist_list # คำสั่ง 'in' บน List คือ O(N)
print(f"List Lookup (O(N)): {time.time() - start:.5f} sec")

# แบบใหม่ (O(1))
start = time.time()
is_fraud_fast = test_account in blacklist_fast # คำสั่ง 'in' บน Dict/Set คือ O(1)
print(f"Fast Lookup (O(1)): {time.time() - start:.5f} sec")
```
