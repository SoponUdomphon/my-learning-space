## **🛠️ Phase 1.3: Clean Code & Refactoring (The "Design")**

มาถึงส่วนที่สำคัญที่สุดในการก้าวขึ้นเป็น Senior Engineer ครับ
อย่างที่เราคุยกันว่า "โค้ดถูกเขียนให้คอมพิวเตอร์เข้าใจแค่ครั้งเดียว แต่ถูกเขียนให้ 'คน' อ่านนับพันครั้ง" ไฟล์ Refactoring and Design Patterns ที่คุณโสภณแนบมาคือคัมภีร์ของเรื่องนี้เลยครับ

ที่ KBTG หรือ SCG คุณต้องทำงานร่วมกับ Engineer อีกหลายสิบคน ถ้าเราเขียนโค้ดที่อ่านยาก (ถึงแม้จะรันเร็วมาก) เวลาเกิดบั๊กตอนตี 2 ระบบอาจจะดาวน์ยาวเพราะเพื่อนร่วมทีมแก้โค้ดเราไม่ได้ครับ

```python
# ❌ The "Arrow Anti-Pattern" (โค้ดทรงหัวลูกศร / Nested If)
def approve_credit_card(user):
    if user.age >= 20:
        if user.salary >= 15000:
            if user.credit_score == "Good":
                if not user.has_criminal_record:
                    return True
                else:
                    return False
            else:
                return False
        else:
            return False
    else:
        return False
```

```python
def approve_credit_card(user):
    if user.age < 20:
        return False

    if user.salary < 15000:
        return False

    if user.credit_score != "Good":
        return False

    if user.has_criminal_record:
        return False

    return True
```
